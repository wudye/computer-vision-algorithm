```python
import numpy as np
```

```python
def dft_slow(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)
```

```python
x = np.random.random(1024)
np.allclose(dft_slow(x), np.fft.fft(x))
```

```python
%timeit dft_slow(x)
%timeit np.fft.fft(x)
```

```python
def fft(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]

    if N % 2 > 0:
        raise ValueError("size of x must be a power of 2")
    elif N <= 32:
        return dft_slow(x)
    else:
        x_even = fft(x[::2])
        x_odd = fft(x[1::2])
        factor = np.exp(-2j * np.pi * np.arange(N) / N)
        half = N // 2
        return np.concatenate([x_even + factor[:half] * x_odd,
                               x_even + factor[half:] * x_odd])
```

```python
x = np.random.random(1024)
np.allclose(fft(x), np.fft.fft(x))
```

```python
%timeit dft_slow(x)
%timeit fft(x)
%timeit np.fft.fft(x)
```

```python
import numpy as np
def fft_vector(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]

    if np.log2(N) % 1 > 0:
        raise ValueError("size of x must be a power 2")

    N_min = min(N, 32)

    n =  np.arange(N_min)
    k = n[:, None]
    M = np.exp(-2j * np.pi * n * k / N_min)
    X = np.dot(M, x.reshape((N_min, -1)))

    while X.shape[0] < N:
        x_even = X[:, : X.shape[1] // 2]
        x_odd = X[:, X.shape[1] // 2:]
        factor = np.exp(-1j * np.pi * np.arange(X.shape[0])
                        / X.shape[0])[:, None]
        X = np.vstack([x_even + factor * x_odd,
                       x_even - factor * x_odd])
    return X.ravel()
```

```python
x = np.random.random(1024)
np.allclose(fft_vector(x), np.fft.fft(x))
```
